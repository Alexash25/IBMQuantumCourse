```python
# Variance of Pauli Measurements in Python

import numpy as np

ket_0 = np.array([1, 0], dtype=complex)

X = np.array([
    [0, 1],
    [1, 0]
], dtype=complex)

Z = np.array([
    [1, 0],
    [0, -1]
], dtype=complex)

def expectation(state, operator):
    # this is <phi|A|phi>
    return np.vdot(state, operator @ state)

def variance(state, operator):
    mean = expectation(state, operator)
    mean_square = expectation(state, operator @ operator)
    return mean_square - mean**2

print("<Z> =", expectation(ket_0, Z))
print("Var(Z) =", variance(ket_0, Z))

print("<X> =", expectation(ket_0, X))
print("Var(X) =", variance(ket_0, X))
```

```python
# Measurement Order in Qiskit

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

simulator = AerSimulator()

# Prepare |0> and measure in Z basis
qc_z = QuantumCircuit(1, 1)
qc_z.measure(0, 0)

result_z = simulator.run(qc_z, shots=1000).result()
print("Z-basis measurement:", result_z.get_counts())

# Prepare |0> and measure in X basis
# Measuring in X basis = apply H, then measure in Z basis
qc_x = QuantumCircuit(1, 1)
qc_x.h(0)
qc_x.measure(0, 0)

result_x = simulator.run(qc_x, shots=1000).result()
print("X-basis measurement:", result_x.get_counts())
```

```python
# Create the commutators for all Pauli Pairs

# Step 1: Define the matrices

X = np.array([
    [0, 1],
    [1, 0]
], dtype=complex)

Y = np.array([
    [0, -1j],
    [1j, 0]
], dtype=complex)

Z = np.array([
    [1, 0],
    [0, -1]
], dtype=complex)

# Step 2: Make a function that will perform the commutator analysis
def commutator(gate1, gate2):
    return (gate1 @ gate2 - gate2 @ gate1)

# Step 3: print the results
print(f"[X, Y] = {commutator(X, Y)}")
print(f"[Y, Z] = {commutator(Y, Z)}")
print(f"[X, Z] = {commutator(Z, X)}")
```
