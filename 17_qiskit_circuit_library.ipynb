```python
# July 15, 2026

# Qiskit Circuit Library

from qiskit import QuantumCircuit
from qiskit.circuit.library import MCXGate

qc = QuantumCircuit(5)

# Multi-controlled X gate with 4 control qubits and 1 target qubit
mcx = MCXGate(4)

# controls: qubits 0, 1, 2, 3. Target: qubit 4
qc.append(mcx, [0, 1, 2, 3, 4])

qc.draw("mpl")
```

```python
# compose() - used when you want to place an entire subcircuit inside a larger circuit

from qiskit import QuantumCircuit

subcircuit = QuantumCircuit(2, name="bell_prep")
subcircuit.h(0)
subcircuit.cx(0, 1)

qc = QuantumCircuit(2)
qc.compose(subcircuit, qubits=[0, 1], inplace=True)
qc.measure_all()

qc.draw("mpl")
```
