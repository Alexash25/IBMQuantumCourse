```python
# June 26, 2026
# 04 Linear Operators and Observables

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

# Expectation Value of Z in Python
import numpy as np

# define amplitudes for |psi> = alpha|0> + beta|1>
alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)

# make column vector input 
psi = np.array([alpha, beta], dtype=complex)

# make the z gate
Z = np.array([
    [1,0],
    [0,-1]
], dtype=complex)

# @ is matrix multiplication
expectation_Z = np.vdot(psi, Z @ psi)

# expectation is 0 bc average cancels out
print(f"<Z> = {expectation_Z}")
```

```python
# Pauli Operators in Qiskit

# make quantum circuit
qc = QuantumCircuit(1)

# apply hadamard
qc.h(0)

# get state
state = Statevector.from_instruction(qc)

# get Z gate (with coefficient 1.0)
Z = SparsePauliOp.from_list([("Z", 1.0)])

# expectation
expectation_Z = state.expectation_value(Z)

print(f"Statevector: {state}")
print(f"<Z> = {expectation_Z}")
```

```python
# Proving [X,Z] do not commute

X = SparsePauliOp.from_list([("X", 1.0)])
Z = SparsePauliOp.from_list([("Z", 1.0)])

# perform commutator
result = X @ Z - Z @ X

print(result)

X = np.array([
    [0,1],
    [1,0]],
    dtype=complex
)

Z = np.array([
    [1,0],
    [0,-1]],
    dtype=complex
)

# perform commutator
result = X @ Z - Z @ X
print(result)
```
