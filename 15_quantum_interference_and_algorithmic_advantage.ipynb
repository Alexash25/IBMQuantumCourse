```python
# July 12, 2026

# Quantum Interference and Algorithmic Advantage

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)

qc.h(0)
qc.h(0)

state = Statevector.from_instruction(qc)

print(state)
```

```python
# Grover Diffusion Operator

from qiskit import QuantumCircuit

qc = QuantumCircuit(2)

# Create uniform superposition
qc.h([0, 1])

# Diffusion operator
qc.h([0, 1])
qc.x([0, 1])
qc.cz(0, 1)
qc.x([0, 1])
qc.h([0, 1])

print(qc.draw())

qc.draw("mpl")
```
