```python
# June 23, 2026
# 03 State Vectors and Dirac Notation

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Create a one-qubit circuit
qc = QuantumCircuit(1)

# Apply H gate
qc.h(0)

# Get resulting statevector
state = Statevector.from_instruction(qc)

print(state)
```

```python
# probabilities
import numpy as np

alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)

p0 = abs(alpha) ** 2
p1 = abs(beta) ** 2

print(f"P(0) = {p0}")
print(f"P(1) = {p1}")
print(f"Total probability = {p0 + p1}")
```

```python
# inner product in python
ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)

inner_01 = np.vdot(ket_0, ket_1)
inner_00 = np.vdot(ket_0, ket_0)

print(f"<0|1> = {inner_01}")
print(f"<0|0> = {inner_00}")
```

```python
# Normalized state with alpha = 0.5
alpha1 = 0.5
beta1 = np.sqrt(3) / 2

p2 = abs(alpha1) ** 2
p3 = abs(beta1) ** 2

print(f"P2 = {p2}")
print(f"P3 = {p3}")
```
