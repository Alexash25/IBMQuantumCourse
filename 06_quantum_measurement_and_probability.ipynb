```python
# July 1, 2026
# 06 Quantum Measurement and Probability

# Simulating Measurement Probabilities in Python
import numpy as np

# complex amplitudes
alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)

# get probabilities
p0 = abs(alpha) ** 2
p1 = abs(beta) ** 2

# num shots
shots = 1000

samples = np.random.choice(
    ["0", "1"],
    size=shots,
    p=[p0, p1]
)

count_0 = np.sum(samples =="0")
count_1 = np.sum(samples =="1")

print(f"Theoretical P(0): {p0}")
print(f"Theoretical P(1): {p1}")
print("Measured counts:")
print(f"0: {count_0}")
print(f"1: {count_1}")
```

```python
# Prepare the state |+> and measure in the computational basis

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# circuit with one qubit and one classical bit
qc = QuantumCircuit(1, 1)

# prepare |+>
qc.h(0)

# measure qubit 0 into classical bit 0
qc.measure(0, 0)

simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts()

print(counts)
```
