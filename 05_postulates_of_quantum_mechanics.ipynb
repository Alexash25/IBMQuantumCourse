```python
# June 27, 2026
# 05 Postulates of Quantum Mechanics

# Born Rule in Python
import numpy as np

alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)

psi = np.array([alpha, beta], dtype=complex)

# Projection operator |0><0|
P0 = np.array([
    [1, 0],
    [0, 0]
], dtype=complex)

# Projection operator |1><1|
P1 = np.array([
    [0, 0],
    [0, 1]
], dtype=complex)

prob_0 = np.vdot(psi, P0 @ psi)
prob_1 = np.vdot(psi, P1 @ psi)

print(f"P(0) = {prob_0.real}")
print(f"P(1) = {prob_1.real}")
print(f"Total = {(prob_0 + prob_1).real}")
```

```python
# Creating a Bell State in Qiskit
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)

qc.h(0)
qc.cx(0, 1)

state = Statevector.from_instruction(qc)

print(state)
```
