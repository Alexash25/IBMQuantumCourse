```python
# 08 Unitary Evolution and Schrodinger Equation

# Checking unitarity

import numpy as np

# build the hadamard
H = (1 / np.sqrt(2)) * np.array([
    [1, 1],
    [1, -1]
], dtype=complex)

# "eye" is used for the identity matrix
identity = np.eye(2)

check = H.conj().T @ H

print(check)
print("Is H unitary?", np.allclose(check, identity))
```

```python
# RZ Gate in Qiskit

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

qc = QuantumCircuit(1)

qc.h(0)
qc.rz(np.pi / 2, 0)

state = Statevector.from_instruction(qc)

print(state)
```
